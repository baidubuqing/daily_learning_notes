#学习使用github的第二天#
git的简单使用：
1.在桌面右键选择更多，选择open git bash here，可以打开git控制台，输入（git clone +复制的github上项目的链接）可以将项目文件直接下载到桌面；
2.在自己电脑桌面新建一个自己的项目的空白文件夹，点进去，右键打开git控制台，输入（git init）进行初始化，会在文件夹里生成.git文件夹来管理你的文件，然后新建代码，它所在的目录也就是工作区，在工作区新建源代码文件，接下来将代码复制到git仓库进行备份，先输入（git add .），此命令是让git将当前文件夹内的所有文件和非空文件夹设置为准备提交状态，这一步不能省；然后输入（git commit -m "备注"）回车后，git会把源代码以数据库的形式保存在仓库中，可以用（git log）查看提交的历史记录;(git checkout HEAD +要恢复的文件)作用是从最后一次的提交里把该文件复制到工作区（会覆盖）；输入(git add +文件名)让该文件单独进入准备提交状态，再输入(git commit -m "备注"),这样每次上传提交的结果就比较清晰。
