```python
# Jupyter可用的交互式计算器（兼容性修正版）
try:
    import ipywidgets as widgets
    from IPython.display import display, clear_output
except ImportError:
    print("请先安装ipywidgets库！")
else:
    output = widgets.Output()
    expr_input = widgets.Text(
        value='',
        placeholder='如2+3*4',
        description='表达式:',
        disabled=False
    )
    button = widgets.Button(description="计算")

    def on_button_clicked(b):
        with output:
            clear_output()
            try:
                result = eval(expr_input.value)
                print(f"结果: {result}")
            except Exception as e:
                print(f"输入有误: {e}")

    button.on_click(on_button_clicked)
    display(expr_input, button, output)
    print("如果没有看到输入框，请确保已安装ipywidgets，并在菜单栏点击‘重启内核并重新运行所有单元格’。")
```
